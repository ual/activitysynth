```python
import orca
import pandas as pd
from urbansim_templates import modelmanager
import random
from urbansim_templates.models import OLSRegressionStep
```

```python
modelmanager.initialize()
```

```python
#Creating random table for learning purposes
df = pd.DataFrame({'sale_price': [random.randint(100000,5000000) for x in range(1000)],
                   'residential_sqft': [random.randint(200,5000) for x in range(1000)] })

orca.add_table('buildings', df)

df_to_predict = pd.DataFrame({'residential_sqft': [random.randint(200,5000) for x in range(1000)]})
orca.add_table('to_predcit',df_to_predict )
```

```python
m = OLSRegressionStep()
m.name = 'price-prediction'
m.tables = 'buildings'
m.model_expression = 'sale_price ~ residential_sqft'
m.out_tables = 'to_predcit'
m.out_column = 'quiro que se guarde en esta columna'
```

```python
m.fit()
```

```python
modelmanager.register(m)
```

```python
orca.run(['price-prediction'])
```

```python
orca.get_table('to_predcit').to_frame()
```

```python
m.run()
```
